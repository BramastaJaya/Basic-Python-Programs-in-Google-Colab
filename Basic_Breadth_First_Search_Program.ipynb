```python
import networkx as nx
import matplotlib.pyplot as plot
from collections import deque

def show_graph(G, pos = None, title=''):
    if pos == None:
        pos = nx.spring_layout(G)

    nx.draw(G, pos, with_labels = True, node_color = 'white', node_size = 100, font_color = 'red', font_weight = 'bold', width = 2)
    label = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels = label, font_color = 'black', font_weight = 'bold', font_size = 10)
    plot.margins(0.1)
    plot.title(title)
    plot.show()

def bfs_path(graph, start, end):
    visited = set()
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == end:
            return path

        if node not in visited:
            visited.add(node)

            for neighbor in graph.neighbors(node):
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)

    return None




G = nx.Graph()
pos = {
    'Jakarta': (0, 4),
    'Bandung': (0, 2.5),
    'Cirebon': (2.5, 3.5),
    'Yogyakarta': (4, 0.5),
    'Semarang': (4, 2.8),
    'Surakarta': (6.2, 1.6),
    'Surabaya': (8.2, 2.8),
    'Malang': (8.5, 0.5)
}

edges = [
    ('Jakarta', 'Bandung', 270),
    ('Jakarta', 'Cirebon', 327),
    ('Bandung', 'Cirebon', 120),
    ('Bandung', 'Yogyakarta', 373),
    ('Cirebon', 'Semarang', 305),
    ('Cirebon', 'Yogyakarta', 210),
    ('Semarang', 'Yogyakarta', 109),
    ('Semarang', 'Surakarta', 97),
    ('Yogyakarta', 'Surakarta', 60),
    ('Semarang', 'Surabaya', 369),
    ('Surabaya', 'Malang', 94),
    ('Surakarta', 'Malang', 370),
]

for u, v, w in edges:
    G.add_edge(u, v, weight=w)

show_graph(G, pos=pos, title='Graf Pulau Jawa')

i = 1
print("urutan yang dikunjungi algoritma Breadth-First-Search: ")

start = 'Bandung'
end   = 'Malang'
path  = bfs_path(G, start, end)

if path:
    print(f"{start} ke {end}: {'  '.join(path)}")
else:
    print("Tidak ditemukan jalur.")

```
